```python
# Convert OrderDate to datetime type.
df['OrderDate'] = pd.to_datetime(df['OrderDate'], errors='coerce')
display(df.info())
```

```python
# Add a new column UnitTotal = Quantity * Price and compare with TotalAmount. Flag mismatches.
df['Price'] = df['Price'].astype(str).str.replace('$', '', regex=False)
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')
df['Unit_total'] = df['Quantity'] * df['Price']
display(df['Unit_total'].head())
```

```python
# Create a new column DiscountApplied (True if UnitTotal > TotalAmount).
df['TotalAmount'] = pd.to_numeric(df['TotalAmount'], errors='coerce')
df['DiscountApplied'] = df['Unit_total'] > df['TotalAmount']
display(df.head())
```

```python
# 24.Display all customers who bought both "Phone" and "Laptop".
phone_buyers = set(df[df['Product'] == 'Phone']['CustomerID'].unique())
laptop_buyers = set(df[df['Product'] == 'Laptop']['CustomerID'].unique())

common_customers = phone_buyers.intersection(laptop_buyers)

both_products_customers_df = df[df['CustomerID'].isin(common_customers)]

unique_customer_names = both_products_customers_df['CustomerName'].unique()

print("Customers who bought both Phone and Laptop:")
for name in unique_customer_names:
    print(name)
```
